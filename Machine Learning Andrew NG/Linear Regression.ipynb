```python
def cost(x, y, theta):
    # 1/2m E( (ho(x, theta) - y)**2)
    # ho(x, theta) = theta.T * x
    
    ho = theta[0] + theta[1] * x
    s = np.sum((ho - y)**2)
    
    
    return s / (2*x.size)
```

```python
cost(x, y, [-2, 0.9])
```

```python
def gradient_descent(x, y, alfa, theta):
        
    ho = theta[0] + theta[1] * x

    s0 = np.sum(ho - y)
    s1 = np.sum((ho-y) * x)

    temp1 = theta[0] - (alfa / x.size) * s0
    temp2 = theta[1] - (alfa / x.size) * s1
    
    while abs(theta[0] - temp1) > 10**(-3) or abs(theta[1] - temp2) > 10**(-3):
        
        theta[0] = temp1
        theta[1] = temp2
    
        ho = theta[0] + theta[1] * x

        s0 = np.sum(ho - y)
        s1 = np.sum((ho-y) * x)

        temp1 = theta[0] - (alfa / x.size) * s0
        temp2 = theta[1] - (alfa / x.size) * s1

        print(f"theta0 = {theta[0]} \t theta1 = {theta[1]} \t cost = {cost(x, y, theta)}")
        
    return theta
    
```

```python
new_theta = gradient_descent(x, y, 0.001, [-2, 0.9])
```
